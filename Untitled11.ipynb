# **graphs:**

```
#line plot graph
a=ts['prices.amountMax']
b=ts['prices.amountMin']
plt.plot(a,label='line one',color='r',marker='o',markerfacecolor='r')
plt.plot(b,label='line two',color='g',marker='o',markerfacecolor='g')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.legend()
plt.title('line plot')
plt.grid(True,color='k',linestyle='--')
plt.show()
```

```
# histogram plot
histogram graph
a=ts['prices.amountMax']
c=range(1,100)
plt.hist(a,c,label='bar one',rwidth=0.8,color='g')
plt.title('histogram')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.legend(loc='upper right')
plt.show()
```

```
#scatter plot
a=ts['prices.amountMax']
b=ts['prices.amountMin']
plt.scatter(a,b,label='line one',color='r',marker='o')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.legend(loc='upper right')
plt.title('scatter plot')
plt.grid(True,color='k',linestyle='--')
plt.show()
```
